# Beautiful Soup 基本用法

```python

# 引入 Beautiful Soup 模組
from bs4 import BeautifulSoup as bs

# 原始 HTML 程式碼 假設我們已經有檔案了
html_doc = """
<html><head><title>前進吧！高捷少女</title></head>
<body><h2>K.R.T. GIRLS</h2>
<p>小穹</p>
<p>艾米莉亞</p>
<p>婕兒</p>
<p>耐耐</p>
<a id="link1" href="https://zh.wikipedia.org/wiki/%E9%AB%98%E6%8D%B7%E5%B0%91%E5%A5%B3#%E5%B0%8F%E7%A9%B9">Link 1</a>
<a id="link2" href="https://zh.wikipedia.org/wiki/%E9%AB%98%E6%8D%B7%E5%B0%91%E5%A5%B3#%E8%89%BE%E7%B1%B3%E8%8E%89%E4%BA%9E">Link 2</a>
<a id="link3" href="https://zh.wikipedia.org/wiki/%E9%AB%98%E6%8D%B7%E5%B0%91%E5%A5%B3#%E5%A9%95%E5%85%92 3</a>
<a id="link4" href="https://zh.wikipedia.org/wiki/%E9%AB%98%E6%8D%B7%E5%B0%91%E5%A5%B3#%E8%80%90%E8%80%90">Link 4</a>
</body></html>
"""

# 以 Beautiful Soup 解析 HTML 程式碼
soup = bs(html_doc, 'html.parser') #前項"html_doc"為html的文字資料，後項"'html.parser'"為指定以何種解析器來分析html文字

print(soup)
```

```python
# 取得節點文字內容

# 獲取標籤內容 : name參數方法，可查找所有名為name的tag。
web_title = soup.title  #取得網頁標題
print(web_title)
print(web_title.string) #使用字串存取
print('\n')

# 找查元素 : 選取全部符合條件(標籤節點)的元素 find_all()
my_girls = soup.find_all('p') #<p></p>標籤
print(my_girls)
print('\n')


# 選取第一個符合條件(標籤節點)的元素 find
my_girls = soup.find('p')
print(my_girls)
print('\n')


# keyword 參數 : 返回具有keyword的元素
my_link = soup.find(id='link1')
print(my_link)
print('\n')
```

# 抓取forbes富豪排行榜

- [2022 The Richest People In The World](https://www.forbes.com/billionaires/)
- [Youtube : How to scrape real time Forbes world billionaires data](https://www.youtube.com/watch?app=desktop&v=iErMVPVun3A)

```python
! pip install pandas
```

```python
import requests
import pandas as pd
# import pprint

url = 'https://www.forbes.com/billionaires/page-data/index/page-data.json'

r = requests.get(url)
# print(r.json())
# pprint --- 数据美化输出
# pp = pprint.PrettyPrinter(indent=2)
# pp.pprint(r.json())


# 抓取需要的資料
data = []
json_data = r.json()
for item in json_data['result']['pageContext']['tableData']:
    try:
        rank = item['rank']
    except:
        rank = ''
    try:
        name = item['personName']
    except:
        name = ''
    try:
        age = item['age']
    except:
        age = ''
    try:
        country = item['country']
    except:
        country = ''
    try:
        source = item['source']
    except:
        source = ''
    try:
        industries = item['industries']
    except:
        industries = ''
    try:
        organization = item['organization']
    except:
        organization = ''
    try:
        selfMade = item['selfMade']
    except:
        rank = ''
    try:
        bios = item['bios']
    except:
        bios = ''
    try:
        abouts = item['abouts']
    except:
        abouts = ''
    try:
        netWorth = item['netWorth']
    except:
        netWorth = ''

    dic = {
        'rank': rank,
        'name': name,
        'age': age,
        'country': country,
        'source': source,
        'industries': industries,
        'organization': organization,
        'selfMade': selfMade,
        'bios': bios,
        'abouts': abouts,
        'netWorth': netWorth,
    }
    data.append(dic)

df = pd.DataFrame(data)
# 印出表格
df


# 產出csv
# df.to_csv('forbes_billionaires_rank.csv', index = False, encoding="utf-8")
# print('download csv finished')
```

```python
# 依照產業分組與個數
df.groupby('industries').size().sort_values(ascending=False)
```

```python
# 選出台灣的富豪
filt = (df['country'] == 'Taiwan')
df.loc[filt]
```

```python
# 選出台灣的富豪再依照產業分組與個數
filt = (df['country'] == 'Taiwan')
df.loc[filt].groupby('industries').size().sort_values(ascending=False)
```
